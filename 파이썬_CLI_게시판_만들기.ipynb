```python
from datetime import datetime

def bulletin():

  print("==CLI게시판 실행==")

  i = 0
  article_list = []

  while True:
    cmd = input("명령어) ")
    if cmd == "exit":
      break

    elif cmd == "article write":
      i += 1
      now  = datetime.now()
      formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
      title = input("제목: ")
      content = input("내용: ")
      print(f"{i}번 글이 생성되었습니다.")
      article_list.append([i, formatted_date, title, content])

    elif cmd.startswith("article delete"):
      num = int(cmd.split()[-1])

      found = False # 못찾았다를 가정해 플래그로 세워둔다.

      for article in article_list:
        if article[0] == num:
          article_list.remove(article)
          print(f"{num}번 글이 삭제되었습니다.")
          found = True # 진짜 찾았다.
          break

      if found == False: # 순회했지만, 아직 True가 된 적이 없다.
        print(f"{num}번 글은 존재하지 않습니다.")

    elif cmd.startswith("article edit"):
      num = int(cmd.split()[-1])

      for article in article_list:
        if article[0] == num:
          # 기존 제목 내용 보여주기
          print(f"기존 제목 : {article[2]}")
          print(f"기존 내용 : {article[3]}")
          # 새로 내용 입력받아서 기존 내용 업데이트
          edit_title = input("수정할 제목: ")
          edit_content = input("수정할 내용: ")
          article[2] = edit_title
          article[3] = edit_content
          print(f"{num}번 글이 수정되었습니다.")
          break
      else:
        print(f"{num}번 글은 존재하지 않습니다.")

    elif cmd.startswith("article detail"):
      num = int(cmd.split()[-1])

      for article in article_list:
        if article[0] == num:
          # 자세히 보여주기
          print(f"번호 : {article[0]}")
          print(f"제목 : {article[1]}")
          print(f"내용 : {article[2]}")
          print(f"날짜 : {article[3]}") # 나중에 인덱스를 추가할 때 순서 바꾸기가 귀찮으니.. key, value 값이 있는 딕셔너리로 하자.

          break
      else:
        print(f"{num}번 글은 존재하지 않습니다.")


    elif cmd == "article list":
      if not article_list:
        print("작성된 글이 없습니다.")
      else:
          print("=============================================")
          print("번호 / 날짜 / 제목 / 내용")
          article_newest = article_list[::-1]
          for article in article_newest:
            print(f"{article[0]} / {article[1]} / {article[2]} / {article[3]} ")
          print("=============================================")
    else:
      print("지원하지 않는 기능입니다.")

  print("==CLI게시판 종료==")

bulletin()

```
