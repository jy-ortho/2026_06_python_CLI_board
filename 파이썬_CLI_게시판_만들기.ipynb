== CLI 게시판 종료==



명령어) article delete "3" , article delete "1"
3번 글은 존재하지 않습니다.  1번 글이 삭제되었습니다

게시글 수정 기능 구현.
게시글

```python
def bulletin():

  print("==CLI게시판 실행==")

  i = 0
  article_list = []

  while True:
    cmd = input("명령어) ")
    if cmd == "exit":
      break

    elif cmd == "article write":
      i += 1
      title = input("제목: ")
      content = input("내용: ")
      print(f"{i}번 글이 생성되었습니다.")
      article_list.append([i, title, content])

    elif cmd.startswith("article delete"):
      num = int(cmd.split()[-1])
      for article_delete in article_list:
        if article_delete[0] == num:
          article_list.remove(article_delete)
          print(f"{num}번 글이 삭제되었습니다.")
        else:
          print(f"{num}번 글은 존재하지 않습니다.")

    elif cmd == "article list":
      if not article_list:
        print("작성된 글이 없습니다.")
      else:
          print("=============================================")
          print("번호 / 제목 / 내용")
          article_newest = article_list[::-1]
          for article in article_newest:
            print(f"{article[0]} / {article[1]} / {article[2]}")
          print("=============================================")
    else:
      print("지원하지 않는 기능입니다.")

  print("==CLI게시판 종료==")

bulletin()

```
