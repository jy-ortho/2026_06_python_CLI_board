== CLI 게시판 실행 ==


명령어) article write

제목: (사용자 입력)

내용: (사용자 입력)

1번 글이 생성 되었습니다.

명령어) article write

제목: (사용자 입력)

내용: (사용자 입력)

2번 글이 생성 되었습니다.

명령어) exit

== CLI 게시판 종료==

```python
import numpy as np

def bulletin():
  print("==CLI게시판 실행==")

  i = 0
  article_list = []

  while True:
    cmd = input("명령어) ")
    if cmd == "exit":
      break

    elif cmd == "article write":
      i += 1
      title = input("제목: ")
      content = input("내용: ")
      print(f"{i}번 글이 생성되었습니다.")
      article_list.append([i, title, content])
      article = np.array(article_list)

    elif cmd == "article list":
      print("=============================================")
      print(article)
      print("=============================================")

    else:
      print("지원하지 않는 기능입니다")

  print("==CLI게시판 종료==")

bulletin()

```
